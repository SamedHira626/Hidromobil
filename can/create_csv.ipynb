```python
import csv
import os
import pandas as pd
import cv2
```

```python
DATADIR = 'TrafficSignDatasets'

with open('veri_onizleme.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['width', 'height', 'class', 'path'])
    
    for i, category in enumerate(os.listdir(DATADIR)):
        path = os.path.join(DATADIR, category)
        for img_name in os.listdir(path):
            
            img_path = os.path.join(path, img_name)
            img = cv2.imread(img_path, 0)
            w, h = img.shape

            writer.writerow([w, h, i, img_path])
```
